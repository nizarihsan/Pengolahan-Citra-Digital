```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca citra grayscale
image = cv2.imread('elang.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Histogram Citra

# Menampilkan citra asli
plt.figure(figsize=(6, 6))
plt.imshow(image, cmap='gray')
plt.title("Citra Grayscale")
plt.axis('off')
plt.show()

# Menghitung histogram
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Normalisasi histogram
hist_normalized = hist / hist.sum()

# Visualisasi histogram asli
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(hist, color='blue')
plt.title("Histogram Asli")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi")
plt.grid()

# Visualisasi histogram normalisasi
plt.subplot(1, 2, 2)
plt.plot(hist_normalized, color='green')
plt.title("Histogram Normalisasi")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi Normalisasi")
plt.grid()

plt.tight_layout()
plt.show()

# Penjelasan hasil
print("""
Histogram menggambarkan distribusi intensitas piksel pada citra grayscale:
1. Sumbu X: Nilai intensitas piksel (0–255).
2. Sumbu Y: Frekuensi kemunculan setiap intensitas.
3. Normalisasi histogram memastikan nilai berada dalam rentang [0, 1], mempermudah analisis distribusi relatif.
""")

# Penjelasan hasil
print("""
Penjelasan Langkah-Langkah:
1. Membaca Citra:
   -Gunakan OpenCV untuk membaca gambar dalam mode grayscale agar histogram memiliki distribusi intensitas tunggal.
2. Menghitung Histogram:
   -Fungsi cv2.calcHist() menghitung histogram untuk satu channel dengan rentang nilai [0, 256].
3. Normalisasi Histogram:
   -Normalisasi dilakukan dengan membagi setiap nilai histogram dengan jumlah total nilai histogram, sehingga hasilnya berada dalam rentang [0, 1].
4. Visualisasi:
   -Plot histogram asli dan hasil normalisasi menggunakan Matplotlib.
   -Berikan label sumbu dan judul untuk memahami maknanya.
5. Penjelasan Hasil:
   -Histogram menunjukkan bagaimana intensitas piksel tersebar dalam citra.
   -Distribusi yang merata berarti citra memiliki kontras yang baik, sementara distribusi terkonsentrasi menunjukkan area yang dominan terang/gelap.
""")
```

```python
# Operasi Konvolusi dengan Kernel Sederhana

# Menampilkan citra asli
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Citra Asli")
plt.axis('off')

# Membuat kernel rata-rata 3x3
kernel = np.ones((3, 3), np.float32) / 9

# Melakukan konvolusi dengan kernel
convolved_image = cv2.filter2D(image, -1, kernel)

# Menampilkan hasil konvolusi
plt.subplot(1, 2, 2)
plt.imshow(convolved_image, cmap='gray')
plt.title("Citra Setelah Konvolusi (Kernel 3x3)")
plt.axis('off')

plt.tight_layout()
plt.show()

# Penjelasan Efek
print("""
Efek Konvolusi dengan Kernel Rata-Rata 3x3:
1. Citra menjadi lebih halus karena setiap piksel digantikan dengan rata-rata intensitas
   dari piksel di sekitarnya (area 3x3).
2. Detail halus pada citra, seperti garis tipis atau noise kecil, berkurang atau menghilang.
3. Efek ini berguna untuk pre-processing data, seperti mengurangi noise sebelum
   analisis lebih lanjut.
""")

# Penjelasan Langkah-Langkah
print("""
Penjelasan Langkah-Langkah:
1. Import Library:
   -Gunakan OpenCV untuk membaca citra dan melakukan operasi konvolusi.
   -Gunakan Matplotlib untuk menampilkan gambar asli dan hasil konvolusi.
2. Membaca Citra:
   -Citra akan diubah menjadi grayscale untuk menyederhanakan proses.
3. Membuat Kernel Rata-Rata 3×3:
   -Kernel rata-rata adalah matriks berisi nilai 1/9 yang digunakan untuk menghitung rata-rata intensitas piksel di sekitar setiap piksel.
4. Melakukan Operasi Konvolusi:
   -Gunakan fungsi cv2.filter2D() untuk menerapkan kernel pada citra.
5. Visualisasi Hasil:
   -Tampilkan citra asli dan citra hasil konvolusi menggunakan Matplotlib.
6. Penjelasan Efek:
   -Efek konvolusi dengan kernel rata-rata menghasilkan citra yang lebih halus, mengurangi noise, dan kehilangan beberapa detail.
""")
```

```python
# Transformasi Fourier

# Menampilkan citra asli
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Citra Asli")
plt.axis('off')

# Melakukan Transformasi Fourier
dft = cv2.dft(np.float32(image), flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)  # Shift zero frequency ke tengah

# Menghitung Magnitude Spectrum
magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))

# Menampilkan Magnitude Spectrum
plt.subplot(1, 3, 2)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title("Magnitude Spectrum")
plt.axis('off')

# Melakukan Invers Transformasi Fourier
dft_ishift = np.fft.ifftshift(dft_shift)
image_reconstructed = cv2.idft(dft_ishift)
image_reconstructed = cv2.magnitude(image_reconstructed[:, :, 0], image_reconstructed[:, :, 1])

# Menampilkan hasil invers
plt.subplot(1, 3, 3)
plt.imshow(image_reconstructed, cmap='gray')
plt.title("Hasil Invers Transformasi")
plt.axis('off')

plt.tight_layout()
plt.show()

# Penjelasan Magnitude Spectrum
print("""
Informasi dari Magnitude Spectrum:
1. Komponen frekuensi rendah (intensitas tinggi di pusat) menggambarkan area halus atau latar belakang dalam citra.
2. Komponen frekuensi tinggi (intensitas di pinggir) menggambarkan detail, seperti tepi atau tekstur dalam citra.
3. Distribusi energi dalam Magnitude Spectrum memberikan wawasan tentang kompleksitas dan pola dalam citra.
""")

# Penjelasan Langkah-Langkah
print("""
Penjelasan Langkah-Langkah:
1. Import Library:
   -Gunakan OpenCV untuk membaca citra dan melakukan Transformasi Fourier.
   -Gunakan Matplotlib untuk visualisasi.
2. Membaca Citra:
   -Ubah citra menjadi grayscale untuk menyederhanakan proses.
3. Melakukan Transformasi Fourier:
   -Gunakan fungsi cv2.dft() untuk menghitung Transformasi Fourier.
   -Tampilkan Magnitude Spectrum untuk melihat distribusi frekuensi dalam citra.
4. Melakukan Invers Transformasi Fourier:
   -unakan fungsi cv2.idft() untuk mengembalikan citra dari domain frekuensi ke domain spasial.
5. Visualisasi Hasil:
   -Tampilkan citra asli, Magnitude Spectrum, dan hasil invers Transformasi Fourier menggunakan Matplotlib.
6. Penjelasan Magnitude Spectrum:
   -Magnitude Spectrum menggambarkan komponen frekuensi tinggi (tepi, detail) dan frekuensi rendah (area halus, latar belakang) dalam citra.
""")
```

```python
# Operasi Aritmatika Citra - Pencerahan

# Menampilkan citra asli
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Citra Asli")
plt.axis('off')

# Menambahkan konstanta untuk pencerahan
brightness_constant = 50
brightened_image = cv2.add(image, np.ones_like(image) * brightness_constant)

# Melakukan clipping nilai piksel
brightened_image_clipped = np.clip(brightened_image, 0, 255).astype(np.uint8)

# Menampilkan citra hasil pencerahan
plt.subplot(1, 2, 2)
plt.imshow(brightened_image_clipped, cmap='gray')
plt.title("Citra Setelah Pencerahan")
plt.axis('off')

plt.tight_layout()
plt.show()

# Penjelasan Perubahan
print(f"""
Pencerahan dilakukan dengan menambahkan konstanta {brightness_constant} ke setiap piksel citra.
Perubahan yang terjadi:
1. Intensitas piksel meningkat, membuat citra tampak lebih terang.
2. Operasi clipping memastikan nilai piksel tidak melebihi rentang [0, 255].
3. Area gelap menjadi lebih terang, sementara area yang sudah terang bisa mencapai nilai maksimum (saturasi).
""")

# Penjelasan Langkah-Langkah
print("""
Penjelasan Langkah-Langkah:
1. Import Library:
   -Gunakan OpenCV untuk membaca citra dan melakukan operasi aritmatika.
   -Gunakan Matplotlib untuk menampilkan hasil pencerahan.
2. Membaca Citra:
   -Citra akan diubah menjadi grayscale untuk memudahkan analisis.
3. Menambahkan Konstanta:
   -Konstanta akan ditambahkan ke semua piksel dalam citra menggunakan operasi aritmatika.
4. Clipping Nilai Piksel:
   -Pastikan nilai piksel tetap berada dalam rentang valid [0, 255] menggunakan np.clip().
5. Visualisasi Hasil:
   -Bandingkan citra asli dengan citra setelah pencerahan menggunakan Matplotlib.
6. Penjelasan Perubahan:
   -Pencerahan citra meningkatkan intensitas piksel, menghasilkan citra yang lebih terang.
""")
```
